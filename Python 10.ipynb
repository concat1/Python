Задание 1

Скачайте с сайта grouplens.org...movielens/ датасет любого размера. Определите, какому фильму было выставлено больше всего оценок 5.0.

Скачен дата сет (ds) ml-25m.zip

```python
import pandas as pd
```

```python
movies=pd.read_csv('movies.csv')
movies[['movieId','title']].head()
```

```python
ratings=pd.read_csv('ratings.csv')
ratings=ratings[['movieId','rating']].head()
```

```python
joined = ratings.merge(movies, on='movieId', how='left')
joined[['movieId','rating','title']].head()
```

```python
grade =ratings['rating'].max()
filtered_joined=joined[(joined ['rating']==grade)]
fr=filtered_joined.groupby(['movieId' , 'title'])['movieId'].count()
```

```python
fr.sort_values(ascending=False).head()
```

Задание 2

По данным файла power.csv посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 года. Не учитывайте в расчетах отрицательные значения quantity.

```python
power=pd.read_csv('power.csv')
power.head()
```

```python
filtered_countries = power[(power['country']=='Estonia') | (power['country']=='Lithuania') | (power['country']=='Latvia') ]
filtered_countries.head()
```

```python
fc=filtered_countries[(filtered_countries['category']==4) | 
                      (filtered_countries['category']==12) | 
                      (filtered_countries['category']==21) & 
                      (filtered_countries['year']<2010) & 
                      (filtered_countries['year']>2005)]
```

```python
print('Итог равен {}КВт'.format(fc['quantity'].sum()))
```
