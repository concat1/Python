Задание 1

Для датафрейма log из материалов занятия создайте столбец source_type по следующим правилам:

если источник traffic_source равен yandex или google, то в source_type ставится organic

для источников paid и email из России - ставим ad

для источников paid и email не из России - ставим other

все остальные варианты берем из traffic_source без изменений

```python
import pandas as pd
```

```python
df_log=pd.read_csv('visit_log.csv', sep=';')
df_log.head()
```

```python
df_log.loc[(df_log.traffic_source == 'yandex') | (df_log.traffic_source == 'google'), 'source_type'] = 'organic'
df_log.loc[((df_log.traffic_source == 'paid') | (df_log.traffic_source == 'email')) & (df_log.region=='Russia'), 'source_type'] = 'ad'
df_log.loc[((df_log.traffic_source == 'paid') | (df_log.traffic_source == 'email')) & (df_log.region!='Russia'), 'source_type'] = 'other'
df_log.loc[df_log.source_type.isnull(), 'source_type']=df_log['traffic_source']
```

Задание 3

Используйте файл с оценками фильмов ml-latest-small/ratings.csv. Посчитайте среднее время жизни пользователей, которые выставили более 100 оценок. Под временем жизни понимается разница между максимальным и минимальным значением столбца timestamp для данного значения userId.

```python
df_ratings=pd.read_csv('ratings.csv', sep=',')
df_ratings.head()
```

```python
df_ratings_count=df_ratings.groupby('userId').count().reset_index()
df_ratings_count=df_ratings_count[['userId','rating']]
df_ratings_count.head()
```

```python
count_more_than_100=df_ratings_count[df_ratings_count['rating']>100]
count_more_than_100.rename(columns={'rating': 'rating_count'}, inplace=True)
count_more_than_100.head()
```

```python
rating_max=df_ratings.groupby('userId').max().reset_index()
rating_min=df_ratings.groupby('userId').min().reset_index()
```

```python
rating_max_timestamp=rating_max[['userId','timestamp']]
rating_min_timestamp=rating_min[['userId','timestamp']]
```

```python
rating_max_timestamp.rename(columns={'timestamp': 'timestamp_max'}, inplace=True)
rating_min_timestamp.rename(columns={'timestamp': 'timestamp_min'}, inplace=True)
```

```python
rating_max_timestamp.head()
```

```python
rating_min_timestamp.head()
```

```python
avg_timestamp=count_more_than_100.merge(rating_max_timestamp, how='inner').merge(rating_min_timestamp, how='inner')
```

```python
def average(row):
    mean_=(row['timestamp_max']-row['timestamp_min'])
    return mean_
```

```python
avg_timestamp['avg']=avg_timestamp.apply(average, axis=1)
len(count_more_than_100)==len(avg_timestamp)
```
